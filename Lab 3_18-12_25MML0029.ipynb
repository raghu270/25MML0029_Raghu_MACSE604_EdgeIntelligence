```python
data['@timestamp'] = (
    data['@timestamp']
    .str.replace(r'(\d+)(st|nd|rd|th)', r'\1', regex=True)
)
data['@timestamp'] = pd.to_datetime(
    data['@timestamp'],
    format='%B %d %Y, %H:%M:%S.%f'
)
```

```python
data.sort_values(['ip_address', '@timestamp'], inplace=True)
```

```python
data.head()
```

```python
data['shift_time'] = data.groupby(['ip_address'])['@timestamp'].shift(1)
```

```python
data.head()
```

```python
data['time_diff'] = (data['@timestamp']-data['shift_time']).dt.seconds//60
```

```python
data.head()
```

```python
data['date'] = data['@timestamp'].dt.date
```

```python
data['weekday'] = data['@timestamp'].dt.weekday
```

```python
data['hour'] = data['@timestamp'].dt.hour
```

```python
data['is_weekend'] = ((data['weekday'] == 5) | (data['weekday'] == 6)).astype(int)
```

```python
data.head()
```

```python
data['hour_bucket'] = data['hour']//4
# This integer division by 4, so we can analyse 4-hour buckets instead of every single hour
```

```python
ip_addr ='ip_address'
```

```python
ip_counts = data.groupby(ip_addr)['@timestamp'].count().reset_index()
```

```python
ip_counts.head()
```

```python
ip_counts = ip_counts.rename(columns={'@timestamp':'total_count'})
```

```python
daily_counts = data.groupby([ip_addr,'date'])['@timestamp'].count().reset_index()
```

```python
daily_counts = daily_counts.rename(columns={'@timestamp':'daily_counts'})
```

```python
daily_counts_avg = daily_counts.groupby(ip_addr).daily_counts.median().reset_index()
```

```python
daily_counts_avg.head(5)
```

```python
weekend_counts = data.groupby([ip_addr, 'is_weekend'])['@timestamp'].count().reset_index()
```
