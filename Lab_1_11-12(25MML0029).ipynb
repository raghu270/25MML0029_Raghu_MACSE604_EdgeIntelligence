# Images Visualization

```python
from PIL import Image
import matplotlib.pyplot as plt
import os
import random

folder = "celeba_images/img_align_celeba/img_align_celeba"

# get list of all images
files = [f for f in os.listdir(folder) if f.endswith(".jpg")]
print("Total images:", len(files))

# choose 5 random images
random_files = random.sample(files, 5)

for filename in random_files:
    img_path = os.path.join(folder, filename)
    img = Image.open(img_path)

    plt.figure(figsize=(3,3))
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"{filename}\nSize: {img.size}")
    plt.show()
```

# Checking the shape of the random images

```python
from PIL import Image
import os
import random

folder = "celeba_images/img_align_celeba/img_align_celeba"

# list all .jpg images
files = [f for f in os.listdir(folder) if f.endswith(".jpg")]

# pick 100 random images
random_files = random.sample(files, 100)

sizes = []

for filename in random_files:
    img_path = os.path.join(folder, filename)
    img = Image.open(img_path)
    
    sizes.append((filename, img.size))  # store filename + (width, height)

# Print results
for name, size in sizes:
    print(f"{name} --> Size: {size}")

print("\nTotal checked:", len(sizes))
```

# checking every image is of the same shape

```python
from PIL import Image
import os

folder = "celeba_images/img_align_celeba/img_align_celeba"

files = [f for f in os.listdir(folder) if f.endswith(".jpg")]

unique_shapes = set()

for filename in files:
    img_path = os.path.join(folder, filename)
    img = Image.open(img_path)
    unique_shapes.add(img.size)   # (width, height)

# Print result
print("Number of unique image sizes:", len(unique_shapes))
print("\nUnique shapes found:")
for shape in unique_shapes:
    print(shape)

# Final verdict
if len(unique_shapes) == 1:
    print("\n✅ All images have the SAME shape.")
else:
    print("\n⚠️ Images have DIFFERENT shapes.")
```
