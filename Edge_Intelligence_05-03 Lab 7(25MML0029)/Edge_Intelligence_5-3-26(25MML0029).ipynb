# Display Video Properties

```python
import cv2

# Open laptop camera
cap = cv2.VideoCapture(0)

# Get frame width and height
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('sample.mp4', fourcc, 20.0, (width, height))

while True:
    ret, frame = cap.read()

    if not ret:
        print("Failed to grab frame")
        break

    # Write frame to video file
    out.write(frame)

    # Show camera
    cv2.imshow("Recording Camera", frame)

    # Press q to stop recording
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

# GrayScale Video

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow("Grayscale video",gray)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Pixel values of the First Frame

```python
import cv2

# open video
cap = cv2.VideoCapture("sample.mp4")

# read first frame
ret, frame = cap.read()

if ret:
    
    # print pixel values of the frame
    print("Pixel values of original frame:")
    print(frame)

    # convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # print grayscale pixel values
    print("Pixel values of grayscale frame:")
    print(gray)

    # display images
    cv2.imshow("Original Frame", frame)
    cv2.imshow("Grayscale Frame", gray)

    cv2.waitKey(0)

cap.release()
cv2.destroyAllWindows()
```

# Shape Drawing

```python
import cv2

# Open webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    if not ret:
        print("Failed to capture frame")
        break

    # Draw rectangle
    cv2.rectangle(frame, (100, 100), (300, 300), (0, 255, 0), 3)

    # Draw circle
    cv2.circle(frame, (200, 200), 50, (255, 0, 0), 3)

    # Add text
    cv2.putText(frame, "OpenCV Practice",
                (50, 50),
                cv2.FONT_HERSHEY_SIMPLEX,
                1,
                (0, 0, 255),
                2)

    # Show video
    cv2.imshow("Shapes", frame)

    # Press q to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release camera and close windows
cap.release()
cv2.destroyAllWindows()
```

# Canny Edge Detection

```python
import cv2 

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    edges = cv2.Canny(frame, 100, 200)   # detects the edges

    cv2.imshow("Edges", edges)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Negative Image

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    negative = 255 - frame

    cv2.imshow("Original", frame)
    cv2.imshow("Negative", negative)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# MOTION DETECTION

```python
import cv2

cap = cv2.VideoCapture(0)

ret, frame1 = cap.read()
ret, frame2 = cap.read()

while True:

    diff = cv2.absdiff(frame1, frame2)
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)

    blur = cv2.GaussianBlur(gray, (5,5), 0)

    _, thresh = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)

    cv2.imshow("Motion", thresh)

    frame1 = frame2
    ret, frame2 = cap.read()

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
