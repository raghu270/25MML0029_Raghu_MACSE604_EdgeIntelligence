```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# 1. Load the model
model = YOLO("yolov8n.pt")

# 2. Open your video
video_path = "/content/Traffic.55.33 PM.mp4"
cap = cv2.VideoCapture(video_path)

# Get properties to ensure the output matches the input exactly
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# 3. Setup VideoWriter
# 'mp4v' or 'avc1' are standard for .mp4 files
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("output_video.mp4", fourcc, fps, (w, h))

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Run tracking (stream=True is more memory efficient for videos)
    results = model.track(frame, persist=True, stream=True)

    for r in results:
        # Plot boxes and IDs on the frame
        annotated_frame = r.plot()

        # Write the processed frame to the video file
        out.write(annotated_frame)

cap.release()
out.release()
cv2.destroyAllWindows()
```
